```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/betulcetin/phishing-url-analysis/refs/heads/main/data/dataset_phishing.csv'
df = pd.read_csv(url)
```

```python
df.describe() # İstatistiksel özet
```

```python
sns.countplot(x='status', data=df) # Hedef Sınıf Dağılımı (Phishing vs Legit)
plt.title("Phishing (1) vs Not Phishing (0)")
plt.show()
```

```python
df.groupby('status')['length_url'].agg(['mean', 'median', 'std']) # URL uzunluk grupla&özetle
```

```python
df['url_length'] = df['url'].apply(lambda x: len(x))

plt.figure(figsize=(10,5))
sns.histplot(df['url_length'], bins=50, kde=True)
plt.title("URL Uzunluk Dağılımı")
plt.xlabel("URL Length")
plt.show()
```

```python
df['has_https'] = df['url'].apply(lambda x: 1 if x.startswith("https://") else 0) # HTTPS kullanımı & Phishing URL’lerin HTTPS oranı

plt.figure(figsize=(6,4))
sns.countplot(x='has_https', hue='status', data=df)
plt.title("HTTPS Kullanımı vs Phishing")
plt.legend(["Not Phishing", "Phishing"])
plt.show()
```

```python
pd.crosstab(df['has_https'], df['status'], normalize='columns').mul(100).round(1) # HTTPS analizi, crosstab + yüzde hesaplama
```
