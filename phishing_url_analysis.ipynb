```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/betulcetin/phishing-url-analysis/refs/heads/main/data/dataset_phishing.csv'
df = pd.read_csv(url)
```

```python
sns.countplot(x='status', data=df) # Hedef Sınıf Dağılımı (Phishing vs Legit)
plt.title("Phishing (1) vs Not Phishing (0)")
plt.show()
```

```python
df['url_length'] = df['url'].apply(lambda x: len(x))

plt.figure(figsize=(10,5))
sns.histplot(df['url_length'], bins=50, kde=True)
plt.title("URL Uzunluk Dağılımı")
plt.xlabel("URL Length")
plt.show()
```
